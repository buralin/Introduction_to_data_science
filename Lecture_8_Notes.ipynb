```python
fp = 'data/mbl.csv'
fp2 = 'data/mbl2.csv'

if not os.path.exists(fp):
    tables = pd.read_html("http://www.baseball-reference.com/teams/BOS/2016-schedule-scores.shtml#team_schedule::none")
    games = tables[0]
    games.to_csv(fp)
else:
    games = pd.read_csv(fp)

games2 = games.loc[games['Gm#'] != 'Gm#']
games2.to_csv(fp2)


column_names = {'Gm#':'game-id' ,'Date': 'date', 'Unamed: 2': 'box', 'Tm': 'box', 'Opp' : 'oponent', 
                'W/L': 'w/l', 'R':'R', 'RA':'RA', 'W-L':'w-l'}

games2 = (games2.rename(columns=column_names)
    [['date', 'oponent', 'w/l', 'w-l']]
#    .assign(date=lambda x: pd.to_datetime(x['date'], format='%A, %b %d'))
    .set_index('date', append=True)
    .rename_axis(["game_id", "date"])
    .sort_index())
          
games2
```
